一、斑马问题

```python
from kanren import *
from kanren.core import lall
def left(x,y,units):
    group = zip(units,units[1:])
    return membero((x,y),group)
def right(x,y,units):
    return left(y,x,units)
def beside(x,y,units):
    return conde([right(x,y,units)],[left(x,y,units)])
house = var()
rule_restrictions = lall(
    (eq,(var(),var(),var(),var(),var()),house),
    #(country, work, drink, pet, color)
    (membero,(var(), var(), var(),"zebra", var()), house),
    (membero,(var(), var(), "water", var(), var()), house),
    #英国人住在红色的房子里
    (membero,("England",var(),var(),var(),"red"),house),
    #西班牙人养了一条狗
    (membero,("Spain",var(),var(),"dog",var()),house),
    #日本人是一个油漆工
    (membero,("Japan","painter",var(),var(),var()),house),
    #意大利人喜欢喝茶
    (membero,("Italy",var(),"tea",var(),var()),house),
    #挪威人住在左边的第一个房子里
    (eq, (("Norway",var(),var(),var(),var()), var(), var(), var(), var()), house),
    # #绿房子在白房子的右边
    (right,(var(),var(),var(),var(),"green"),(var(),var(),var(),var(),"white"),house),
    # #摄影师养了一只蜗牛
    (membero,(var(),"photographer",var(),"snails",var()),house),
    # #外交官住在黄房子里
    (membero,(var(),"diplomat",var(),var(),"yellow"),house),
    # #中间那个房子的人喜欢喝牛奶
    (eq,(var(),var(),(var(),var(),"milk",var(),var()),var(),var()),house),
    # #喜欢喝咖啡的人住在绿房子
    (membero,(var(),var(),"coffee",var(),"green"),house),
    # 挪威人住在蓝色的房子旁边
    (beside, ("Norway", var(), var(), var(), var()), (var(), var(), var(), var(), "blue"),
     house),
    # # 小提琴家喜欢喝橘子汁
    (membero,(var(),"violinist","juice",var(),var()),house),
    # #养狐狸的人所住的房子与医生的房子相邻
    (beside,(var(),var(),var(),"fox",var()),(var(),"doctor",var(),var(),var()),house),
    # #养马的人所住的房子与外交官的房子相邻
    (beside,(var(), var(), var(), "horse", var()), (var(), "diplomat", var(), var(), var()), house)
)
solutions = run(1,house,rule_restrictions)
print(solutions)
house1 = [house for house in solutions[0] if "zebra" in house][0][4]
print("{}房子里的人养斑马".format(house1))
house2 = [house for house in solutions[0] if "water" in house][0][4]
print("{}房子里的人喝水".format(house2))
```

二、八皇后问题

```python
def conflict(state, nextColumn):
    nextRow = rows = len(state) # 5
    for row in range(rows):   #  0,1,2,3,4
        # 获取当前行的列
        column = state[row]
        if abs(column - nextColumn) in [0, nextRow - row]:
            return True
    return False
def queens(num, state=()):

    # 每一行的列坐标都是从0:7的
    # 0,1,2,3,4,5,6,7
    for pos in range(num):
        if not conflict(state, pos): # 回溯法的体现

            if len(state) == num - 1:
                # 最后一行的（pos,）=最后一行的result，然后再递归回去求倒数第二行的result
                yield (pos,)
            else:
                for result in queens(num, state + (pos,)):
                    yield (pos,) + result

if __name__ == '__main__':
    solutions = queens(8)
    for index, solution in enumerate(solutions):
        print(solution )
```
